```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split as tts
from sklearn.ensemble import RandomForestRegressor
import matplotlib.pyplot as ppt
```

```python
data = pd.read_csv("data2pred.csv")
x = data.iloc[:,1:-1].values
y = data.iloc[:,-1].values
print(x)
```

```python
print(y)
```

```python
ppt.title("Year vs Birth Rates, Duties, Urban Rate, and Foreigner Percentage")
ppt.xlabel("Year")
ppt.ylabel("Crude Birth Rate / Duties / Urban Rate\nand Foreigner Percentage")
years = data.iloc[:,0].values

ppt.plot(years,x[:,0],color='blue')
ppt.plot(years,x[:,2],color='red')
ppt.plot(years,x[:,4],color='cyan')
ppt.plot(years,y,color='green')
ppt.legend(['Duty Percent','Foreign Born Percent','Urban Rate','Birth Rate'])
ppt.savefig('initial.png')
```

```python
xtn,xtt,ytn,ytt = tts(x,y,test_size=0.25,random_state=9)
print(xtn)
```

```python
model = RandomForestRegressor(n_estimators=149,max_depth=149,random_state=9)
model.fit(xtn,ytn)
print(model.score(xtt,ytt))
```

```python
predic = pd.read_csv("crystal.csv")
realBirths = pd.read_csv("realBirths.csv")
xpred = predic.iloc[:,1:].values
years2 = predic.iloc[:,0:1].values
births = realBirths.iloc[:,1:].values
print(xpred)
```

```python
print(years2)
```

```python
print(births)
```

```python
ypred = model.predict(xpred)
print(model.score(xpred,ypred))
```

```python
print(ypred)
```

```python
ppt.clf()
ppt.title("Year vs Predicted and Real Birth Rate")
ppt.xlabel("Year")
ppt.ylabel("Birth Rates")

ppt.plot(years2,ypred,color='Red')
ppt.plot(years2,births,color='green')
ppt.savefig('prediction.png')
```
